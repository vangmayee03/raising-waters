```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_excel("dataset/flood dataset.xlsx")
```

```python
dataset.isnull().any()
```

```python
print(dataset.shape)
print(dataset.columns)
print(dataset.dtypes)
print(dataset.describe())
```

```python
x = dataset[['Cloud Cover','ANNUAL','Jan-Feb','Mar-May','Jun-Sep']]
```

```python
y = dataset.iloc[:,10].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=10)
```

```python
from sklearn import tree
from sklearn import ensemble
from sklearn import neighbors
from xgboost import XGBClassifier

# Define models with controlled complexity

dtree = tree.DecisionTreeClassifier(
    
)

rf = ensemble.RandomForestClassifier()

knn = neighbors.KNeighborsClassifier()

xgb = XGBClassifier()
```

```python
dtree.fit(x_train, y_train)
rf.fit(x_train, y_train)
knn.fit(x_train, y_train)
xgb.fit(x_train, y_train)
```

```python
p1  = dtree.predict(x_test)
p2 = rf.predict(x_test)
p3 = knn.predict(x_test)
p4 = xgb.predict(x_test)
```

```python
from sklearn import metrics

print(metrics.accuracy_score(y_test,p1))
print(metrics.accuracy_score(y_test, p2))
print(metrics.accuracy_score(y_test, p3))
print(metrics.accuracy_score(y_test, p4))

```

```python
metrics.confusion_matrix(y_test,p4)
```
